```python
# Question 3: Drop redundant columns

fifa_df = fifa_df.drop(columns =['Photo','Flag','Club Logo'])
```

```python
# Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix.

selection = fifa_df[['Value','Wage','Release Clause']]
print(selection)
```

```python
## let us replace '€' as '' in the column value and convert "K" and "M" for thousand and million value
fifa_df['Wage'] = fifa_df['Wage'].str.replace('€','')
fifa_df['Wage'] = (fifa_df['Wage'].replace(r'[KM]+$', '', regex=True).astype(float) * 
             fifa_df['Wage'].str.extract(r'[\d\.]+([KM]+)', expand=False)
                       .fillna(1)
                       .replace(['K','M'],[10**3,10**6]).astype(float))
fifa_df.head()
```

```python
fifa_df[['Value','Wage','Release Clause']]
```

```python
# convert the column "joined" into integer data type with keeping only year.

fifa_df['Joined']
```

```python
fifa_df['Joined'] = fifa_df['Joined'].str.split(',', expand=True)[1]
fifa_df['Joined'] = pd.to_numeric(fifa_df['Joined'].str.strip())
fifa_df['Joined'] = pd.to_numeric(fifa_df['Joined'], errors='coerce').fillna(0).astype(int)
fifa_df['Joined']
```

```python
# Convert the column "Contract Valid Until" to pandas datetime type.

fifa_df.loc[:,"Contract Valid Until"]
```

```python
# convert the contract valid until to datetime type
fifa_df['Contract Valid Until'] = pd.to_datetime(fifa_df['Contract Valid Until'])
fifa_df.loc[:,"Contract Valid Until"]
```

```python
# The column 'Height' is in inches with a quotation mark, Convert to float with decimal points.

fifa_df['Height'] = fifa_df['Height'].str.split("'", expand=True)[0] + '.' + fifa_df['Height'].str.split("'", expand=True)[1]
fifa_df['Height'] = pd.to_numeric(fifa_df['Height'])
fifa_df['Height']
```

```python
# The column "Weight" has the suffix as lbs, remove the suffix and convert to float.

fifa_df['Weight'] = pd.to_numeric(fifa_df['Weight'].str.replace('lbs',''))
fifa_df['Weight']
```

```python
# Check for the percentage of missing values and impute them with appropriate imputation techniques.

fifa_df.isnull().sum()/len(fifa_df) * 100
```

```python
for col in ['Preferred Foot','International Reputation','Weak Foot','Skill Moves','Work Rate','Body Type','Position','Jersey Number','Contract Valid Until','Height','Weight'
,'Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys','Dribbling','Curve','FKAccuracy','LongPassing','BallControl','Acceleration','SprintSpeed'
,'Agility','Reactions','Balance','ShotPower','Jumping','Stamina','Strength','LongShots','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Marking'
,'StandingTackle','SlidingTackle','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes']:
    fifa_df[col].fillna(fifa_df[col].mode()[0], inplace=True)
```

```python
fifa_df['Club'].fillna("No Club", inplace = True)
fifa_df
```

```python
#  Plot the distribution of Overall rating for all the players and write your findings.

ratings = fifa_df["Overall"]
plt.hist(ratings)
plt.xlabel('Overall Rating')
plt.ylabel('Players')
plt.title('Distribution of Players Overall Ratings')
```

```python
# Retrieve the names of top20 players based on the Overall rating.

top_20_names = fifa_df[['Name','Overall']].sort_values(by = 'Overall', ascending = False).head(20).reset_index ()
top_20_names
```

```python
# Generate a dataframe which should include all the information of the Top 20 players based on the Overall rating.

top_20_players_dtls = fifa_df.sort_values(by = 'Overall', ascending = False).head(20).reset_index()
top_20_players_dtls
```

```python
#  What is the average "Age" and "Wage" of these top 20 players? (use the data frame created in the question 11)

avg_age = top_20_players_dtls['Age'].mean()
avg_wage = top_20_players_dtls['Wage'].mean()

print('Average Age of Top 20 Players is: ',avg_age)
print('Average Wage of Top 20 Players is: ',avg_wage)
```

```python
# Among the top 20 players based on the Overall rating, which player has the highest wage? Display the name of the player with his wage.

max_wage = top_20_players_dtls['Wage'].max()
print('Maximum Wage is: ',max_wage)
max_wage_player = top_20_players_dtls.loc[top_20_players_dtls['Wage'] == max_wage, 'Name']
print('Player Name: ',max_wage_player)
```
